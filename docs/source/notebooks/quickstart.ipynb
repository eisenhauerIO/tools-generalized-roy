# Quickstart

This notebook demonstrates the core functionality of `grmpy`: simulating data and estimating marginal treatment effects.

For the economic theory behind the generalized Roy model, see the [course documentation](https://eisenhauerio.github.io/courses-business-decisions/).

## Setup

````python
import grmpy
````

## 1. Simulate Data

Simulate a dataset from the generalized Roy model using a YAML configuration file.

````python
data = grmpy.simulate("files/tutorial.grmpy.yml")
data.head()
````

The simulated data contains:
- `Y1`, `Y0`: Potential outcomes
- `D`: Treatment indicator
- `Y`: Observed outcome
- `X2`, `X3`: Covariates
- `U1`, `U0`, `V`: Unobservables (available in simulation)

## 2. Parametric Estimation

Estimate the model parameters assuming joint normality of unobservables.

````python
rslt = grmpy.fit("files/tutorial.grmpy.yml", semipar=False)
````

## 3. Semiparametric Estimation

For semiparametric estimation using Local Instrumental Variables (LIV), use:

```python
rslt_semipar = grmpy.fit("config.yml", semipar=True)
```

This approach makes no distributional assumptions on unobservables. See the [tutorial](../tutorial.md) for configuration details.
